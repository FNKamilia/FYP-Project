```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('punkt')
```

```python
# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_text(text):
    # Tokenize the text into individual words
    words = nltk.word_tokenize(text)
    # Lemmatize each word using the WordNetLemmatizer
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    # Join the lemmatized words back into a single string
    lemmatized_text = ' '.join(lemmatized_words)
    return lemmatized_text
```

```python
df3['Clean_Review'] = df3['Clean_Review'].apply(lemmatize_text)
```

The library going to be used is called NLTK (Natural Language Toolkit), and specifically, the SentimentIntensityAnalyzer class from NLTK's nltk.sentiment.vader module.

### Count word frequency

```python
#import counter
from collections import Counter
```

```python
def get_tokens(docx,num=30):
    word_tokens = Counter(docx)
    most_common = word_tokens.most_common(num)
    result = dict(most_common)
    return result
```

#### Consideration whether to do TF-DF first or Lemmatizing first

Yes, it is generally recommended to perform TF-IDF (Term Frequency-Inverse Document Frequency) transformation before lemmatization when conducting topic modeling.

Here's the rationale behind this recommendation:

TF-IDF captures term importance: TF-IDF calculates the importance of terms based on their frequency within individual documents and their rarity across the entire document collection. By applying TF-IDF before lemmatization, you retain the original word forms and their frequencies, allowing the TF-IDF algorithm to consider the specific occurrences of terms. This can be beneficial for topic modeling, as it helps in identifying important and distinctive terms within the document collection.

Lemmatization reduces word variations: Lemmatization is the process of reducing words to their base or dictionary form (lemmas). It helps in reducing word variations and normalizing the text. By performing lemmatization after TF-IDF, you ensure that different inflected forms of a word are transformed to their common base form. This can improve the consistency of term representations and reduce the sparsity of the data, which can be useful for topic modeling algorithms that rely on patterns and co-occurrences of terms.

By following the TF-IDF before lemmatization approach, you take advantage of both the term importance captured by TF-IDF and the normalization benefits of lemmatization. This can lead to more meaningful and coherent topic models.

#### Consideration whether to do Lemmatization or Stemming

For LDA (Latent Dirichlet Allocation) and topic modeling tasks, it is generally recommended to choose either lemmatization or stemming, but not both. Using both techniques together can result in redundant information and potentially lead to overfitting.

As for which technique is better, it depends on the specific requirements of your task and the characteristics of your dataset. Here are some considerations for choosing between lemmatization and stemming:

Lemmatization:

Lemmatization converts words to their base form, considering the part of speech. This can help preserve the semantic meaning of the words, which is important for topic modeling.
Lemmatization tends to produce more meaningful and interpretable topics compared to stemming.
However, lemmatization can be computationally more expensive than stemming.


Stemming:

Stemming reduces words to their root or stem form by removing prefixes or suffixes. It is a simpler and faster process compared to lemmatization.
Stemming may result in words that are not actual words and can sometimes lead to less meaningful topics compared to lemmatization.
Stemming is useful when you want to focus more on the word structure and pattern rather than the semantic meaning.
In most cases, lemmatization is preferred for topic modeling tasks like LDA because it tends to produce more meaningful and interpretable topics. However, it's always a good idea to experiment with both techniques and evaluate their impact on the quality of the topics generated for your specific dataset.

#### Consideration whether to do Lemmatizing or Sentiment Analysis first

It is generally recommended to perform lemmatization before conducting sentiment analysis. Lemmatization helps normalize the text by reducing words to their base or root form, which aids in capturing the underlying meaning of the words. By reducing the variation in word forms, lemmatization improves the accuracy of sentiment analysis models by grouping together different forms of the same word.

When you lemmatize the text first and then analyze the sentiment, you ensure that sentiment patterns are captured consistently across different inflected forms of words. This approach helps in improving the accuracy and reliability of sentiment analysis results.


#### Lemmatizing First, Then Sentiment Analysis:

Pros:

Improved word normalization: Lemmatization reduces words to their base or root form, which helps in normalizing the text. This normalization can improve the accuracy of sentiment analysis by grouping together different forms of the same word.
Consistency in sentiment analysis: By lemmatizing the text first, you ensure that sentiment patterns are captured consistently across different inflected forms of words.
Reduced vocabulary variation: Lemmatization reduces the number of unique words in the text by mapping inflected forms to a common lemma. This can help sentiment analysis models handle variations in vocabulary and improve coverage.
Cons:

Increased preprocessing complexity: Lemmatization adds an extra preprocessing step before conducting sentiment analysis, which may require additional computational resources and time.
Possible loss of word context: In some cases, lemmatizing the text before sentiment analysis may lead to a loss of word context. Certain sentiment nuances can be tied to specific word forms, and lemmatization may not capture these nuances accurately.

#### Sentiment Analysis First, Then Lemmatizing:

Pros:

Simplicity and efficiency: Performing sentiment analysis first allows for a simpler and more efficient workflow, as there is no need for an additional preprocessing step of lemmatization.
Quick sentiment insights: By obtaining sentiment labels or scores early on, you can gain rapid insights into the sentiment of the text without the delay of lemmatizing.
Cons:

Limited word normalization: Without lemmatization, variations in word forms may not be adequately captured, potentially affecting the accuracy and consistency of sentiment analysis.
Reduced coverage and accuracy: Sentiment analysis models may struggle with variations in vocabulary and may not accurately capture sentiment patterns across different inflected forms of words.
