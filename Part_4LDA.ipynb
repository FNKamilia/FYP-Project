```python
#data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd
```

```python
#Loading file excel
df4 = pd.read_excel('LDAwords.xlsx')
```

```python
df4
```

```python
unique_topic_ids = df4['topicId'].unique()

for topic_id in unique_topic_ids:
    words = df4.loc[df4['topicId'] == topic_id, 'word'].tolist()
    print(f"Topic {topic_id}: {words}")
    print("-------------------------------------------------------------------------------------------------------------------------------")
```

# Conclusion

Topic 0: Service
Because there's mentioned about service, staff, friendly

Topic 1: Price /  Value
Because after topic 0, and 2, there's mentioned price right after, reasonable, worth

Topic 2: Food
Because we can see that list of foods


Topic 3: Ambiance
Because a lot of mentioned about surrounding about the restaurant like what the restaurant have like table, restaurant, came, like, place, experience. Attribute that don't exist at other topics, like not a descriptive

```python
#replace values
topic_names = {0: "Service", 1: "Price", 2: "Food", 3: "Ambiance"}

df4['topicId'] = df4['topicId'].replace(topic_names)
```
