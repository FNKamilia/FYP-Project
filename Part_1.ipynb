#### Dropping the unrelevant reviews based on 'Year'

We want to drop some of the reviews that is consider unrelevant to the analysis.

```python
# Count the occurrences of each unique row in the "Year" column
unique_counts = df['Year'].value_counts().reset_index()

# Rename the columns for clarity
unique_counts.columns = ['Year', 'Count']

# Sort the dataframe in descending order by the "Year" column
unique_counts_sorted = unique_counts.sort_values('Year', ascending=False).reset_index(drop=True)

# Display the sorted unique rows and their counts
print(unique_counts_sorted)
```

From Year 2022 to 2008, there is a 14 years gap in between. 
From Year 2008 to Year 2019, it shows an increasing trend while from Year 2019 to Year 2022, it shows an decreasing trend.

These are the analysis that we want to do in order to determine which Year are relevant and unrelevant to our analysis

Recent Years: Analyzing reviews from the most recent years (e.g., 2021 and 2022) will provide insights into the current sentiments and trends in the restaurant industry. This can be particularly valuable if you are interested in understanding the present scenario.

Trend Analysis: Including a range of years, such as the last 5-10 years, can help identify long-term trends and patterns in sentiment. This will allow you to observe changes and shifts in consumer sentiments over time.

Industry Changes: Considering years when significant changes occurred in the industry, such as the rise of food delivery platforms or the impact of the COVID-19 pandemic, can offer valuable insights into how these events affected customer sentiments and behavior.

Sample Size: Assessing the number of reviews available for each year is also important. If certain years have very few reviews compared to others, it may affect the statistical significance and reliability of the sentiment analysis results for those years.

#### Detect and drop review that was not written in English

Here are a few reasons as for why we need to detect and drop review that was not written in English.

1. Language Consistency: This analysis focuses specifically on English text, removing non-English reviews ensures consistency.

2. Language Understanding: When performing text analysis tasks such as sentiment analysis, topic modeling, or natural language processing, it's crucial to work with text in a language that I understand. By removing non-English reviews, I can focus on analyzing and interpreting English text effectively.

3. Quality Control: Non-English reviews may introduce noise or errors in your analysis. By removing them, it can enhance the quality and accuracy of analysis results.

4. Target Audience: This analysis is targeted towards English-speaking users, it's essential to focus on their reviews to gain insights and make informed decisions that cater to their needs.

```python
from langdetect import detect

# Function to detect the language of a text
def detect_language(text):
    try:
        return detect(text)
    except:
        return 'Unknown'

# Apply language detection to the 'Review' column
df['Language'] = df['Review'].apply(detect_language)
```

```python
non_eng = df[df['Language'] != 'en']
print(non_eng['Review'])
```

```python
from googletrans import Translator

def translate_to_english(text):
    translator = Translator()
    try:
        translation = translator.translate(text, dest='en')
        if translation is not None:
            return translation.text
        else:
            return ''
    except Exception as e:
        print(f"Translation Error: {e}")
        return ''


# Create a copy of the non_english_reviews DataFrame
non_eng_copy = non_eng.copy()

# Apply translation to the 'Review' column in the copy DataFrame
non_eng_copy['Translated_Review'] = non_eng_copy['Review'].apply(translate_to_english)

# Display the translated reviews
print(non_eng_copy['Translated_Review'])
```

```python
for index, row in non_eng_copy.iterrows():
    print(f"Review ID: {index}")
    print(row['Translated_Review'])
    print("--------------------")
```

```python
# Transfer the translated reviews to the 'Review' column in the original dataframe
df.loc[non_eng_copy.index, 'Review'] = non_eng_copy['Translated_Review']
```
